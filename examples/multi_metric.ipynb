# Exemplo: Execuções com múltiplas métricas
----------------------------

Este exemplo mostra como avaliar um pipeline do experionml com múltiplas métricas.

Importe o conjunto de dados de câncer de mama de [sklearn.datasets](https://scikit-learn.org/stable/datasets/index.html#wine-dataset). Este é um conjunto de dados pequeno e fácil de treinar cujo objetivo é prever se uma paciente tem câncer de mama ou não.

## Carregar os dados

```python
# Import packages
import pandas as pd
from experionml import ExperionMLRegressor
```

```python
# Carregue os dados
X = pd.read_csv("./datasets/abalone.csv")

# Let's have a look
X.head()
```

## Executar o pipeline

```python
experionml = ExperionMLRegressor(X, n_jobs=1, verbose=2, random_state=1)
```

```python
experionml.encode()
```

```python
# Em cada etapa da BO, ambas as métricas são calculadas,
# mas apenas a primeira é usada para otimização!
experionml.run(
    models=["lsvm", "hGBM"],
    metric=("r2", "rmse"),
    n_trials=10,
    n_bootstrap=6,
)
```

```python
# Verifique a robustez do pipeline usando validação cruzada
experionml.winner.cross_validate()
```

## Analisar os resultados

```python
# As colunas no dataframe de resultados contêm uma para cada métrica
experionml.results.data[["r2_ht", "r2_train", "r2_test", "rmse_ht", "rmse_train", "rmse_test"]]
```

```python
# Alguns gráficos permitem escolher a métrica que queremos mostrar
with experionml.canvas():
    experionml.plot_trials(metric="r2", title="Hyperparameter tuning performance for R2")
    experionml.plot_trials(metric="rmse", title="Hyperparameter tuning performance for RMSE")
```

```python
experionml.plot_results(metric="r2")
```
