# Exemplo: PLN
--------------

Este exemplo mostra como usar o ExperionML para ir rapidamente de dados de texto brutos a predições de modelo.

Importe o conjunto de textos 20 newsgroups de [sklearn.datasets](https://scikit-learn.org/0.19/datasets/twenty_newsgroups.html). O conjunto contém cerca de 18000 artigos sobre 20 tópicos. O objetivo é prever o tópico de cada artigo.

## Carregar os dados

```python
import numpy as np
from experionml import ExperionMLClassifier
from sklearn.datasets import fetch_20newsgroups
```

```python
# Use apenas um subconjunto dos tópicos disponíveis para processamento mais rápido
X_text, y_text = fetch_20newsgroups(
    return_X_y=True,
    categories=[
        'sci.med',
        'comp.windows.x',
        'misc.forsale',
        'rec.autos',
    ],
    shuffle=True,
    random_state=1,
)
X_text = np.array(X_text).reshape(-1, 1)
```

## Executar o pipeline

```python
experionml = ExperionMLClassifier(X_text, y_text, index=True, test_size=0.3, verbose=2, random_state=1)
```

```python
experionml.dataset  # Observe que a variável é nomeada automaticamente como 'corpus'
```

```python
# Vamos dar uma olhada no primeiro documento
experionml.corpus[0]
```

```python
# Limpe os documentos removendo ruído, emails, números etc.
experionml.textclean()
```

```python
# Verifique como o primeiro documento mudou
experionml.corpus[0]
```

```python
# Converta as strings em uma sequência de palavras
experionml.tokenize()
```

```python
# Exiba as primeiras palavras do primeiro documento
experionml.corpus[0][:7]
```

```python
# Normalize o texto para um padrão predefinido
experionml.textnormalize(stopwords="english", lemmatize=True)
```

```python
experionml.corpus[0][:7]  # Check changes...
```

```python
# Visualize as palavras mais comuns com uma nuvem de palavras
experionml.plot_wordcloud(figsize=(700, 500))
```

```python
# Observe os bigramas mais frequentes
experionml.plot_ngrams(2)
```

```python
# Crie os bigramas usando o tokenizador
experionml.tokenize(bigram_freq=215)
```

```python
experionml.bigrams_
```

```python
# Como última etapa antes da modelagem, converta as palavras em vetores
experionml.vectorize(strategy="tfidf")
```

```python
# A dimensionalidade do conjunto de dados aumentou bastante!
experionml.shape
```

```python
# Observe que os dados são esparsos e as colunas são nomeadas
# de acordo com as palavras que elas representam
experionml.dtypes
```

```python
# Quando o conjunto de dados é esparso, stats() mostra a densidade
experionml.stats()
```

```python
# Verifique quais modelos têm suporte a matrizes esparsas
experionml.available_models(accepts_sparse=True)
```

```python
# Treine o modelo
experionml.run(models="RF", metric="f1_weighted")
```

## Analisar os resultados

```python
experionml.evaluate()
```

```python
experionml.plot_confusion_matrix(figsize=(700, 600))
```
