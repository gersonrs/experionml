# Exemplo: Pruning
------------------

Este exemplo mostra um caso avançado de uso de ajuste de hiperparâmetros com pruning.

Importe o conjunto de dados de câncer de mama de [sklearn.datasets](https://scikit-learn.org/stable/datasets/index.html#wine-dataset). Este é um conjunto de dados pequeno e fácil de treinar cujo objetivo é prever se uma paciente tem câncer de mama ou não.

## Carregar os dados

```python
# Import packages
from sklearn.datasets import load_breast_cancer
from optuna.pruners import HyperbandPruner
from experionml import ExperionMLClassifier
```

```python
# Carregar os dados
X, y = load_breast_cancer(return_X_y=True)
```

## Executar o pipeline

```python
# Initialize experionml
experionml = ExperionMLClassifier(X, y, verbose=2, random_state=1)
```

```python
# Use ht_params para especificar um pruner personalizado
# Observe que trials podados mostram o número de iterações que completaram
experionml.run(
    models="SGD",
    metric="f1",
    n_trials=25,
    ht_params={
        "distributions": ["penalty", "max_iter"],
        "pruner": HyperbandPruner(),
    }
)
```

## Analisar os resultados

```python
experionml.plot_trials()
```

```python
experionml.plot_timeline()
```

```python
experionml.plot_hyperparameter_importance()
```
