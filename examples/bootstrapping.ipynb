# Exemplo: Bootstrapping
--------------------

Este exemplo mostra como usar a técnica de [bootstrapping](https://en.wikipedia.org/wiki/Bootstrapping_(statistics)) para avaliar a robustez de um modelo.

Importe o conjunto de dados de câncer de mama de [sklearn.datasets](https://scikit-learn.org/stable/datasets/index.html#wine-dataset). Este é um conjunto de dados pequeno e fácil de treinar cujo objetivo é prever se uma paciente tem câncer de mama ou não.

## Carregar os dados

```python
# Import packages
from sklearn.datasets import load_breast_cancer
from experionml import ExperionMLClassifier
```

```python
# Carregar os dados
X, y = load_breast_cancer(return_X_y=True, as_frame=True)
```

## Executar o pipeline

```python
# Inicialize o experionml e treine vários modelos usando bootstrapping
experionml = ExperionMLClassifier(X, y, verbose=2, random_state=1)
experionml.run(models=["LR", "Tree", "LGB"], n_bootstrap=10)
```

```python
# Mostre os resultados dos scores de bootstrap
experionml.lr.bootstrap
```

```python
experionml.plot_bootstrap()
```
