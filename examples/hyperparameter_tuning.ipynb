# Exemplo: Ajuste de hiperparâmetros
--------------------------------

Este exemplo mostra um caso avançado de como otimizar os hiperparâmetros do seu modelo em execuções com múltiplas métricas.

Importe o conjunto de dados de câncer de mama de [sklearn.datasets](https://scikit-learn.org/stable/datasets/index.html#wine-dataset). Este é um conjunto de dados pequeno e fácil de treinar cujo objetivo é prever se uma paciente tem câncer de mama ou não.

## Carregar os dados

```python
# Import packages
from sklearn.datasets import load_breast_cancer
from optuna.distributions import IntDistribution
from experionml import ExperionMLClassifier
```

```python
# Carregar os dados
X, y = load_breast_cancer(return_X_y=True)
```

## Executar o pipeline

```python
# Initialize experionml
experionml = ExperionMLClassifier(X, y, n_jobs=4, verbose=2, random_state=1)
```

```python
# Treine um modelo MultiLayerPerceptron com duas métricas
# usando um número personalizado de camadas ocultas
experionml.run(
    models="MLP",
    metric=["f1", "ap"],
    n_trials=10,
    est_params={"activation": "relu"},
    ht_params={
        "distributions": {
            "hidden_layer_1": IntDistribution(2, 4),
            "hidden_layer_2": IntDistribution(10, 20),
            "hidden_layer_3": IntDistribution(10, 20),
            "hidden_layer_4": IntDistribution(2, 4),
        }
    },
    errors='raise'
)
```

```python
# Em execuções com múltiplas métricas, o melhor trial selecionado é o primeiro da frente de Pareto
experionml.mlp.best_trial
```

```python
experionml.plot_pareto_front()
```

```python
# Se você não estiver satisfeito com os resultados, é possível continuar o estudo
experionml.mlp.hyperparameter_tuning(n_trials=5)
```

```python
# O atributo trials fornece uma visão geral dos resultados dos trials
experionml.mlp.trials
```

```python
# Select a custom best trial...
experionml.mlp.best_trial = 2

# ...e verificar que os melhores parâmetros agora são os do trial selecionado
experionml.mlp.best_params
```

```python
# Por fim, ajuste o modelo no conjunto de treino completo 
# usando a nova combinação de hiperparâmetros
experionml.mlp.fit()
```
