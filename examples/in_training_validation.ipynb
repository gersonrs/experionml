# Exemplo: Validação durante o treino
---------------------------------

Este exemplo mostra como acompanhar o desempenho do modelo durante o treino.

Importe o conjunto de dados de câncer de mama de [sklearn.datasets](https://scikit-learn.org/stable/datasets/index.html#wine-dataset). Este é um conjunto de dados pequeno e fácil de treinar cujo objetivo é prever se uma paciente tem câncer de mama ou não.

## Carregar os dados

```python
# Import packages
from sklearn.datasets import load_breast_cancer
from experionml import ExperionMLClassifier
```

```python
# Carregar os dados
X, y = load_breast_cancer(return_X_y=True)
```

## Executar o pipeline

```python
# Initialize experionml
experionml = ExperionMLClassifier(X, y, verbose=2, random_state=1)
```

```python
# Nem todos os modelos oferecem suporte a validação durante o treino
# Você pode verificar quais oferecem usando o método available_models
experionml.available_models(validation=True)
```

```python
# Execute os modelos normalmente
experionml.run(models=["MLP", "LGB"], metric="auc")
```

## Analisar os resultados

```python
experionml.plot_evals(title="In-training validation scores")
```

```python
# Plote a validação nos conjuntos de treino e teste
experionml.lgb.plot_evals(dataset="train+test", title="LightGBM's in-training validation")
```
