# Exemplo: Successive halving
-----------------------------

Este exemplo mostra como comparar múltiplos modelos baseados em árvores usando successive halving.

Importe o conjunto de dados California housing de [sklearn.datasets](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.fetch_california_housing.html).
 Este é um conjunto de dados pequeno e fácil de treinar cujo objetivo é prever preços de imóveis.

## Carregar os dados

```python
from sklearn.datasets import fetch_california_housing
from experionml import ExperionMLRegressor
```

```python
# Carregar os dados
X, y = fetch_california_housing(return_X_y=True)
```

## Executar o pipeline

```python
experionml = ExperionMLRegressor(X, y, verbose=2, random_state=1)
```

```python
# Compare modelos baseados em árvores via successive halving
experionml.successive_halving(
    models=["Tree", "Bag", "ET", "RF", "LGB", "CatB"],
    metric="mae",
    n_bootstrap=5,
)
```

## Analisar os resultados

```python
# Os resultados agora são multi-index, em que frac é a fração
# do conjunto de treino usada para ajustar o modelo. Os nomes dos modelos
# terminam com o número de modelos ajustados naquela execução
experionml.results
```

```python
# Plote os resultados do successive halving
experionml.plot_successive_halving()
```

```python
# Use regex para chamar todos os modelos com o mesmo estimador...
experionml.plot_errors(models=["CatB.*"])
```

```python
# ...ou para chamar os modelos da mesma execução
experionml.plot_errors(models=".*3")
```
