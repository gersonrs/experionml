# Exemplo: Conjunto holdout
----------------------

Este exemplo mostra quando e como usar o conjunto holdout do ExperionML em um pipeline de exploração.

Os dados utilizados são uma variação do [conjunto de dados de clima australiano](https://www.kaggle.com/jsphyg/weather-dataset-rattle-package) do Kaggle. Você pode baixá-lo [aqui](https://github.com/tvdboom/ExperionML/blob/master/examples/datasets/weatherAUS.csv). O objetivo deste conjunto é prever se vai chover amanhã treinando um classificador binário com a variável alvo `RainTomorrow`.

## Carregar os dados

```python
# Import packages
import pandas as pd
from experionml import ExperionMLClassifier
```

```python
# Carregue os dados
X = pd.read_csv("./datasets/weatherAUS.csv")

# Let's have a look
X.head()
```

## Executar o pipeline

```python
# Inicialize o experionml especificando uma fração do conjunto de dados para holdout
experionml = ExperionMLClassifier(X, n_rows=0.5, holdout_size=0.2, verbose=2)
```

```python
# As frações de teste e holdout são separadas após a subamostragem do conjunto de dados
# Observe também que o conjunto holdout não faz parte do dataset do experionml
print("Length loaded data:", len(X))
print("Tamanho dataset + holdout:", len(experionml.dataset) + len(experionml.holdout))
```

```python
experionml.impute()
experionml.encode()
```

```python
# Diferentemente de treino e teste, o conjunto holdout não é transformado até ser usado para predições
experionml.holdout
```

```python
experionml.run(models=["GNB", "LR", "RF"])
```

```python
experionml.plot_prc()
```

```python
# Com base nos resultados no conjunto de teste, selecionamos o melhor modelo para ajuste adicional
experionml.run("lr_tuned", n_trials=10)
```

## Analisar os resultados

Já usamos o conjunto de teste para escolher o melhor modelo para ajuste adicional, então esse conjunto deixou de ser verdadeiramente independente. Embora isso talvez não fique visível diretamente nos resultados, usar o conjunto de teste agora para avaliar o modelo LR ajustado seria um erro, pois ele já carrega viés. Por esse motivo, separamos um conjunto extra e independente para validar o modelo final: o holdout. Se não vamos usar o conjunto de teste para validação, também podemos aproveitá-lo para treinar o modelo e otimizar o uso dos dados disponíveis. Use o método full_train para isso.

```python
# Retreine o modelo no conjunto completo, treino + teste 
experionml.lr_tuned.full_train()
```

```python
# Avalie no conjunto holdout
experionml.lr_tuned.evaluate(rows="holdout")
```

```python
experionml.lr_tuned.plot_prc(rows="holdout", legend="upper right")
```
