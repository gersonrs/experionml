# Exemplo: Previsão univariada
--------------------

Este exemplo mostra como usar o ExperionML com um conjunto de dados de séries temporais univariadas.

Importe o conjunto de dados de companhias aéreas de [sktime.datasets](https://www.sktime.net/en/latest/api_reference/auto_generated/sktime.datasets.load_airline.html). Este é um conjunto de dados pequeno e fácil de treinar que mede os totais mensais de passageiros de companhias aéreas internacionais de 1949 a 1960.

## Carregar os dados

```python
# Import packages
import numpy as np
from sktime.datasets import load_airline
from experionml import ExperionMLForecaster
```

```python
# Carregar os dados
y = load_airline()

print(y)
```

## Analisar os dados

```python
experionml = ExperionMLForecaster(y, verbose=2, random_state=1)
```

```python
# Como o conjunto contém apenas a coluna alvo, experionml.X fica vazio
experionml.X
```

```python
# Examinar o conjunto de dados
experionml.plot_series()
```

```python
experionml.plot_qq()
```

## Sazonalidade

```python
# O ExperionML tem vários gráficos para entender a sazonalidade dos dados
with experionml.canvas(rows=2, cols=1, sharex=True, vspace=0.01, legend=None):
    experionml.plot_acf()
    experionml.plot_pacf()
```

```python
with experionml.canvas(rows=2, cols=1, sharex=True, vspace=0.01, legend=None):
    experionml.plot_periodogram()
    experionml.plot_fft()
```

```python
# Também é possível calcular a sazonalidade
sp = experionml.get_seasonal_period()
sp
```

```python
# E usar essa sazonalidade
experionml.sp = sp
```

## Executar o pipeline

```python
# Use os métodos regulares de limpeza de dados para transformar a coluna alvo
experionml.scale(columns=-1)
```

```python
experionml.y
```

```python
experionml.run(["AutoARIMA", "TBATS", "Prophet"])
```

```python
# Também é possível usar modelos de regressão em tarefas de previsão
experionml.run("RF")
```

```python
experionml.rf.estimator
```

## Analisar os resultados

```python
experionml.evaluate()
```

```python
experionml.winner.plot_forecast()
```

```python
experionml.plot_errors()
```
